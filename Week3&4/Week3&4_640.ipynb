```python
world_df = pd.read_excel("world-population.xlsm")
world_df.head()
```

```python
sns.lineplot(data=world_df, x="Year", y="Population").set(title='World Population over the years')
```

```python
sns.lineplot(data=world_df, x="Year", y="Population", 
             drawstyle='steps-pre').set(title='World Population over the years (step function)')
plt.gcf().set_size_inches(10,8)
plt.show()
```
