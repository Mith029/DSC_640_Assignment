```python
# Mithil Patel
# DSC 630 - Week 1 & 2 Assignment
# 12/09/2022

import pandas as pd

hotdog_df = pd.read_excel("hotdog-contest-winners.xlsm")
hotdog_df.head()
```

```python
# Creating a bar chart
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=hotdog_df, x="Winner", y="Dogs eaten").set_title('Average Hotdogs eaten by each winner')
plt.xticks(rotation=70)
plt.tight_layout()
plt.gcf().set_size_inches(10,15)
```

```python
# Importing another dataframe
obama_df = pd.read_excel("obama-approval-ratings.xls")
obama_df.head()
```

```python
# Creating a stacked chart
obama_df.set_index('Issue').plot(kind='bar', stacked=True, color=['red', 'skyblue', 'green'])
 
# labels for x & y axis
plt.xlabel('Issues')
plt.ylabel('Ratings')
 
# title of plot
plt.title("Obama's rating on each issue")
plt.gcf().set_size_inches(10,5)
```

```python
## Creating a pie chart

# Adding up the number of hotdogs eaten by each country
sums = hotdog_df['Dogs eaten'].groupby(hotdog_df.Country).sum()

# selecting Seaborn color palette to use
palette_color = sns.color_palette('bright')

# plotting data
plt.pie(sums, labels=sums.index, colors=palette_color,autopct='%.0f%%')

# title of plot
plt.title("Percentage of hotdogs eaten by winning country (Pie chart)")
plt.gcf().set_size_inches(5,5)

# displaying chart
plt.show()
```

```python
## Creating a donut chart

plt.pie(sums, labels=sums.index, colors=palette_color,autopct='%.0f%%', pctdistance=0.85)

# drawing a white circle
circle = plt.Circle((0,0), 0.60, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(circle)

# title of plot
plt.title("Percentage of hotdogs eaten by winning country (Donut chart)")
plt.gcf().set_size_inches(5,5)

# displaying chart
plt.show()
```
